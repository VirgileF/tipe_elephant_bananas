```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
b = 3000
c = 1000

x = np.arange(0, 10)
f1 = lambda x: (c-2)*x + c-1
f2 = lambda x: b - 2*x - 1

plt.scatter(x, f1(x), c='r', marker='d', label='Capacité globale (c-2)x + c-1')
plt.scatter(x, f2(x), c='b', marker='o', label='Nombre de bananes restant à transporter b-2x-1')
plt.grid()
plt.xlabel('Nombre d\'aller-retours')
plt.ylabel('Nombre de bananes transportées')
plt.legend()
plt.show()
```

```python
##
# P calcule le nombre optimal de bananes au point d’arrivée. 
# B quantite initiale de bananes.
# C capacite maximale de transport de bananes.
# L distance entre le point de depart et d’arrivée.
##

def P(B, C, L):
    while L>0 and B>0:
        if B%C == 1 and B > 1:
            # on doit abandonner une banane sur le chemin. 
            B -= 1
        B = B - 2*((B-2)//C) - 1
        L -= 1 
    return max(B,0)
```

```python
P(3000, 1000, 1000)
```
